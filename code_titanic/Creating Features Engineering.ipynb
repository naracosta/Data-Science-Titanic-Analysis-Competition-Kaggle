## Transform Data

```python
train[['Ticket_1','Ticket_2','Ticket_3']] = train.Ticket.str.split(" ",expand=True,)
train[['Cabin_1','Cabin_2','Cabin_3','Cabin_4']] = train.Cabin.str.split(" ",expand=True,)
train[['Name_first','Name_second']] = train.Name.str.split(",",expand=True,)
train[['Name_1','Name_2','Name_3','Name_4','Name_5','Name_6','Name_7','Name_8','Name_9','Name_10','Name_11','Name_12','Name_13','Name_14']] = train.Name.str.split(" ",expand=True,)
```
